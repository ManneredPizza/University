```python
import math
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

plt.rcParams['text.usetex'] = True
```

```python
def f(theta, Omega, epsilon):
    return (theta+Omega+np.dot(epsilon,np.sin(theta)))%(2*math.pi)
def F(theta, Omega, epsilon, N):
    if N==1:
        return theta+Omega+np.dot(epsilon,np.sin(theta))
    else:
        return F(theta+Omega+np.dot(epsilon,np.sin(theta)), Omega, epsilon, N-1)
```

```python
Om = np.linspace(0,1,100)
th = np.linspace(0,1,100)
```

```python
plt.scatter(th,f(np.dot(2*math.pi,th),0.1*2*math.pi,0)/(2*math.pi),1.2)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f(\theta)$')
plt.show()
```

```python
#w = lim n->+inf (F^n(th)-th)/n
N=1000
t = 2*math.pi*0.8
eps = 0
plt.scatter(Om,(F(t,np.dot(Om,2*math.pi),eps,N)-t)/(2*math.pi*N),1.2)
plt.xlabel(r'$\Omega$')
plt.ylabel(r'$\omega$')
plt.show()
```

```python
plt.scatter(th,f(np.dot(2*math.pi,th),0.1*2*math.pi,1)/(2*math.pi),1.2)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f(\theta)$')
plt.show()
```

```python
#w = lim n->+inf (F^n(th)-th)/n
N=1000
t = 2*math.pi*0.8
eps = 1
plt.scatter(Om,(F(t,np.dot(Om,2*math.pi),eps,N)-t)/(2*math.pi*N),.5)
plt.xlabel(r'$\Omega$')
plt.ylabel(r'$\omega$')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))
t = 2*math.pi*0.8


epsy = np.linspace(0,2,100)
Omx = np.dot(Om,2*math.pi)

omega = np.ones((100,100))

for i in range(100):
    for j in range(100):
        omega[i,j] *= (F(t, Omx[i], epsy[j], N)-t)/(2*math.pi*N)

Omx, epsy = np.meshgrid(Omx, epsy)

ax.set_xlabel(r'$\Omega$')
ax.set_ylabel(r'$\epsilon$')
ax.set_zlabel(r'$\omega$')

ax.plot_surface(Omx, epsy, omega, cmap='viridis', edgecolor='none')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

z = np.ones((100,100))

for i in range(100):
    for j in range(100):
        z[i,j] *= math.exp(-(x[i]**2+y[j]**2))

x,y = np.meshgrid(x,y)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
plt.show()
```
