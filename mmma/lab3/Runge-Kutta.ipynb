```python
import math
import matplotlib.pyplot as plt
```

```python
def f(x,t):
    return x+0.5*t

def runge_kutta(tau, x0, Nmax):
    x = [x0]
    t = [0]
    for i in range(Nmax):
        h = (tau/2)*f(x[i],t[i])
        x.append(x[i]+tau*f(x[i]+h,t[i]+tau/2))
        t.append(t[i]+tau)
    return x,t
```

```python
x,t = runge_kutta(1/1000, 0, 10000)
```

```python
plt.plot(t,x)
```

```python
def result(ts):
    x=[]
    for t in ts:
        x.append(0.5*math.exp(t)-0.5*t-0.5)
    return x
```

```python
plt.plot(t,result(t))
```

```python
tau = 1/100
w = 2
N = 1000
p = [0.2]
q = [0.3]
```

```python
def f_2(x):
    return [x[1], -(w**2)*x[0]]
```

```python
for i in range(N):
    h = [(tau/2)*f_2([q[i],p[i]])[0], (tau/2)*f_2([q[i],p[i]])[1]]
    q.append(q[i]+tau*f_2([q[i]+h[0],p[i]+h[1]])[0])
    p.append(p[i]+tau*f_2([q[i]+h[0],p[i]+h[1]])[1])
```

```python
plt.plot(p,q)
```
