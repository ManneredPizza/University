```python
import math
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def f(l,x,n):
    if(n==1):
        return l*x*(1-x)
    else:
        return f(l,l*x*(1-x),n-1)
```

```python
lTest = np.linspace(0,4,6)
x0 = np.linspace(-0.1,1.1,100)
```

# Mappe Logistiche

```python
fig, ax = plt.subplots(1,figsize=(8,8))
for l in lTest[1:]:
    ax.plot(x0,f(l,x0,1))
ax.plot([-0.2,1],[-0.2,1])
plt.show()
```

# Stampo, fissato x0 e lambda, il percorso

```python
#fisso un lambda ed un x0
fig, ax = plt.subplots(1,figsize=(8,8))
xs = 0.1
l = 3.5
ax.plot(x0,f(l,x0,1))
ax.plot([min(x0), max(x0)], [min(x0), max(x0)])
ax.set_ylim([min(f(l,x0,1)), max(f(l,x0,1))+0.1])
for i in range(20):
    x_new = f(l,xs,1)
    if i==0:
        ax.plot([xs,xs],[xs,x_new],'g')
        ax.plot([xs,x_new],[x_new,x_new],'g')
    else:
        ax.plot([xs,xs],[xs,x_new],'k')
        ax.plot([xs,x_new],[x_new,x_new],'k')
    xs = x_new
plt.show()
```

```python
#fisso un lambda ed un x0
fig, ax = plt.subplots(1,figsize=(8,8))
xs = 0.1
l = 3.5
ax.plot(x0,f(l,x0,2))
ax.plot([min(x0), max(x0)], [min(x0), max(x0)])
ax.set_ylim([min(f(l,x0,2)), max(f(l,x0,2))+0.1])
for i in range(20):
    x_new = f(l,xs,2)
    if i==0:
        ax.plot([xs,xs],[xs,x_new],'g')
        ax.plot([xs,x_new],[x_new,x_new],'g')
    else:
        ax.plot([xs,xs],[xs,x_new],'k')
        ax.plot([xs,x_new],[x_new,x_new],'k')
    xs = x_new
plt.show()
```

# Stampo Successione 300 punti

```python
#fisso un lambda ed un x0
fig, ax = plt.subplots(1,2,figsize=(16,8))
xn = 0.1
l = 1.5
for i in range(300):
    ax[0].plot([i],[xn], marker=".", color='k')
    xn = f(l,xn,1)
ax[1].plot(x0,f(l,x0,1))
ax[1].plot([-0.1,1],[-0.1,1])
ax[1].ylim([-0.2,0.4])
plt.show()
```

# Grafico di Biforcazione

```python
fig, ax = plt.subplots(1,figsize=(16,8))
lSet = np.linspace(0,4,1000)
st = time.time()
for l in lSet:
    x0 = 0.1
    for i in range(1000):
        xNew = f(l,x0,1)
        if(i>700):
            ax.plot([l],[xNew], marker=",", color='k')
        x0 = xNew
ed = time.time()
ax.title.set_text('Biforcazione da l=0 a l=4, tempo esecuzione='+str(ed-st))
plt.show()
```

```python
fig, ax = plt.subplots(1,figsize=(16,8))
lSet = np.linspace(0,4,1000)
st = time.time()
for l in lSet:
    x0 = 0.1
    if l>=3.35 or (l>0.85 and l<1.15) or (l>2.85 and l<3.15):
        for i in range(1000):
            xNew = f(l,x0,1)
            if(i>700):
                ax.plot([l],[xNew], marker=",", color='k')
            x0 = xNew
    else:
        for i in range(50):
            xNew = f(l,x0,1)
            if(i>40):
                ax.plot([l],[xNew], marker=",", color='k')
            x0 = xNew
ed = time.time()
ax.title.set_text('Biforcazione da l=0 a l=4, tempo esecuzione='+str(ed-st))
plt.show()
```
